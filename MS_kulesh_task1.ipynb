```python
par = 5
sample_exp = np.random.exponential(par, 10000)
sample_uni = np.random.uniform(0, par, 10000)
```

```python
warnings.filterwarnings("ignore")
aprs_uni, aprs_exp  = [], []
stds_uni, stds_exp  = [], []
for k in range(1, 1000):
    aprs_uni.append(apr_uni(sample_uni, k))
    stds_uni.append(std(sample_uni, aprs_uni[-1]))
for k in range(1, 150):
    aprs_exp.append(apr_exp(sample_exp, k))
    stds_exp.append(std(sample_exp, aprs_exp[-1]))
```

```python
plt.figure(figsize = (20, 16))
ax1 = plt.subplot(231)
ax2 = plt.subplot(232)
dots_uni = [i for i in range(1, 1000)]
dots_exp = [i for i in range(1, 150)]

ax1.plot(dots_uni, aprs_uni, color="C0", label='ОЦЕНКА РАВНОМЕРНОЕ РАСПРЕДЕЛЕНИЕ ', linewidth=0.5)
ax1.axhline(y=par, ls="--", lw = 1, c = 'black', label = 'РЕАЛЬНЫЙ ПАРАМЕТР')
ax2.plot(dots_exp, aprs_exp, color="C0", label='ОЦЕНКА ПОКАЗАТЕЛЬНОЕ РАСПРЕДЕЛЕНИЕ ', linewidth=0.5)
ax2.axhline(y=par, ls="--", lw = 1, c = 'black', label = 'РЕАЛЬНЫЙ ПАРАМЕТР')
 
ax1.legend(), ax2.legend()
plt.show()
```
